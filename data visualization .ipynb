```python
x  = [1, 2, 3, 4, 5]

plt.plot(x, [i for i in x], label = 'linear')
plt.plot(x, [i ** 2 for i in x], label = 'square')
plt.plot(x, [i ** 3 for i in x], label = 'cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Curves')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]

plt.plot(x, [i for i in x], label = 'linear', linestyle = "--", marker = 'o')
plt.plot(x, [i ** 2 for i in x], label = 'square', linestyle = "-.", marker = '*', color = 'green')
plt.plot(x, [i ** 3 for i in x], label = 'cube', marker = 's', color = 'purple')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Curves')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]

plt.plot(x, [i for i in x], label = 'linear', linestyle = "--", marker = 'o')
plt.plot(x, [i ** 2 for i in x], label = 'square', linestyle = "-.", marker = '*', color = 'green')
plt.plot(x, [i ** 3 for i in x], label = 'cube', marker = 's', color = 'purple')

for i, (xi, yi) in enumerate(zip(x, [i ** 3 for i in x])):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords = 'offset points', xytext=(-10, 10), ha='center')

plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Curves')
plt.legend()
plt.show()
```

```python
data_dict = {'A' : 25, 'B' : 70, 'C' : 55, 'D' : 90}


# This is done since dictionary in python is non-iterable
for i, key in enumerate(data_dict):
    plt.bar(i, data_dict[key])
    
# Set the labels for x-axis
plt.xticks(np.arange(len(data_dict)), data_dict.keys())  # Adding the keys as x labels
    
plt.show()
```

```python
df = pd.read_csv("Iris.csv")

df.head()
```

```python
df.drop(labels = ['Id'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.Species.value_counts()
```

```python
sns.scatterplot(x = df['SepalLengthCm'], y = df['SepalWidthCm'], hue = df['Species'])
plt.legend()
plt.xlabel('Sepal Length (in cm)')
plt.ylabel('Sepal Width (in cm)')
plt.show()
```

```python
sns.boxplot(x = 'Species', y = 'PetalLengthCm', data = df)
```

```python
sns.violinplot(x='Species',y='SepalLengthCm',data=df)
```

```python
sns.pairplot(data=df, kind='scatter', hue = 'Species')
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr(), annot = True)
plt.show()
```
